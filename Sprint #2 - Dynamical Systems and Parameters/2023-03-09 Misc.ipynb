```python
%matplotlib inline
from pylab import *
from pyndamics3 import Simulation
```

```python
sim=Simulation()
sim.add(" x' = v",50)
sim.add(" v' = -g+(k*v**2)/m",(sqrt(3*5/1)))
sim.params(k=1,m=5,g=3)
sim.run(10)

t=sim.t
v=sim.v

figure(figsize=(8,4))
plot(t,v)
plot(t,(sqrt(3*5/1))*ones(len(t)),'r--')
```

```python
sim.v[0],sim.v[-1]
```
