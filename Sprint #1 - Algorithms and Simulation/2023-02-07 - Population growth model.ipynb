$$
\frac{dp}{dt}=a\cdot p\cdot (1-p/K)
$$

$$
\frac{dp}{dt}=a\cdot p\cdot (1-p/K)+B
$$


```python
p=2
t=0

a=1
K=10

dt=0.01

S=Storage()

while t<10:
    
    dp=(a*p*(1-p/K))*dt
    
    p=p+dp
    t=t+dt
    
    S+=t,p
    
    
t,p=S.arrays()
    
    

```

```python
figure(figsize=(10,4))
plot(t,p)
```
