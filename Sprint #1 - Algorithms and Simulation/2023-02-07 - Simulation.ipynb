```python
%matplotlib inline
from pylab import *
from sci378 import *
```

variables - initial values

```python
y=200
v=0
t=0
```

parameters - values

```python
m=1
g=10
```

```python
dt=0.01
```

loop the model equations

$$
\begin{aligned}
\frac{dy}{dt}&=v \\
\frac{dv}{dt}&=-g
\end{aligned}
$$

```python
S=Storage()

S+=t,y,v
while t<10:
    
    dy=v*dt
    dv=-g*dt
    
    v=v+dv
    y=y+dy
    t=t+dt
    
    S+=t,y,v
    
t,y,v=S.arrays()
```

```python
plot(t,y)
ylabel('height')
```

```python
figure(figsize=(8,4))
subplot(2,1,1)
plot(t,y)
ylabel('height')
subplot(2,1,2)
plot(t,v)
ylabel('speed')
```
